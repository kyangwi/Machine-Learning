```python
import cv2
import matplotlib.pyplot as plt

cv2.__version__
```

### Ploting function for images

```python
def imshow(title='image',img=None,size=5):
    # aspect ration allow to maintain the image integrity despite the 
    img_ratio = img.shape[0] / img.shape[1]
    # converting image to rgb
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    # seting the figure parameters
    plt.figure(figsize=(size*img_ratio,size))
    plt.imshow(img)
    plt.title(title)
    plt.show()

img=cv2.imread('title.jpg')
imshow('Happy new year',img=img)
```

### RGB COLOR DECOMPOSITION

```python
import numpy as np
zeros = np.zeros(img.shape[:2],dtype='uint8')
# image decomposition
R,G,B = cv2.split(img)

imshow('Red', cv2.merge([zeros,zeros,R]))
imshow('Red',cv2.merge([zeros,G,zeros]))
imshow('Red',cv2.merge([B,zeros,zeros]))
```

```python
imshow('Fliped image',cv2.flip(img,4))
```

### Drawing some figures

```python
# let's create an image from scratch
image = np.ones((512,512,3),np.uint8)

imshow('Blank image',image)
```

```python
# Drawing a line
img_line = image.copy()
img_line = cv2.line(img_line,(0,0),(512,512),(127,0,127),3)
imshow('Line on image',img_line)
```

```python
# drowing a rectangle
img_bbox = image.copy()
img_bbox = cv2.rectangle(img_bbox,(100,200),(200,300),(127,0,127),3)
imshow('Drawing bounding box', img_bbox)
```

```python
# 'Hello world
new_image = image.copy()
new_image = cv2.putText(new_image, 'Hello world', (80,270), cv2.FONT_HERSHEY_SIMPLEX, 2, (0,127,0), 4)
imshow('Hello world',new_image)
```
